## 1.Reading from CSV file

```python
train_data=pd.read_csv("/kaggle/input/home-data-for-ml-course/train.csv")
test_data=pd.read_csv("/kaggle/input/home-data-for-ml-course/test.csv")
```

```python
k=train_data.isnull().sum()
k.sort_values(ascending=False)[0:25]
```

## 4.Selecting Features for Modelling

```python

features=['LotArea','1stFlrSF','2ndFlrSF','YearBuilt','TotalBsmtSF','BedroomAbvGr','FullBath',
          'GarageArea','GrLivArea','TotRmsAbvGrd',
         'Fireplaces','OverallQual']
X=train_data[features]
X.describe()
```

## 5.Specifying and Fitting Model

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
y=train_data.SalePrice
```

```python
model.fit(X,y)
```

## 6. Reviewing Data for Testing

```python
test_data.head()
```

```python
features
```

```python
k=test_data[features]
k.describe()

```

```python
k.isnull().sum()
```

## 7. Data Wrangling on Test Data

```python
k['TotalBsmtSF'].fillna(k['TotalBsmtSF'].mean(),inplace=True)
k['GarageArea'].fillna(k['GarageArea'].mean(),inplace=True)
```

```python
k.describe()
```

```python
k.isnull().sum()
```

## 8.Predicting the Price

```python
predicted=model.predict(k)
```

## 9.Output to CSV

```python
output=pd.DataFrame({'Id':test_data.Id,'SalePrice':predicted})
output.to_csv('submmision.csv',index=False)
print('successfull')
```
